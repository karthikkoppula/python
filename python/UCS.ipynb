```python
import os
import requests
from pprint import pprint
import json
import re
import pandas as pd
from github import Github
import pdb
import yaml
import plotly.express as px
from datetime import datetime, date, timedelta
from ast import literal_eval
import PyPDF2
```

```python
def laas_yaml():
    laas_excel = "Device_pool_updated.xlsx"
    ucs_info = pd.read_excel(laas_excel, sheet_name="UCS") # ucs sheet full
    #print(ucs_info)
    rack_numbers = ['PP28']

    for rack_number in rack_numbers:
        rack_info = pd.read_excel(laas_excel, sheet_name=rack_number)  
#         print(rack_info)
        routers = [48,56] 
    
        for router in routers:
            try:    
                with open('dev_ENG-RTP-VTST-52_2345(3).yaml', 'r') as file:
                     demo_yaml = yaml.safe_load(file)
#                      print(demo_yaml)
                    
                     
                print("Router number is {}".format(router))
                peer = rack_info['Peer'][router] 
               
                ucs_details = ucs_info[ucs_info['Hostname '] == peer]
                

                cores = ucs_details['Cores'].iloc[0]
                memory = ucs_details['Memory'].iloc[0]
      

                name = 'SDWAN-POOL/SDWAN-VTEST/SDWAN-VTEST-L1SW/SDWAN-VTEST-RACK{}CORE/{}'.format(cores,peer)
                print(name)
                location = 'BGL11/3/{}/{}'.format(rack_number[0]+rack_number[1],rack_number[2]+rack_number[3])
                email = 'eng-bgl-laas-admin@cisco.com'
        
                host_ip = ucs_details['Host IP'].iloc[0].strip()
                netmask = 22
                gateway = '10.104.33.1'
                tags = [cores,memory,peer] 
                
                
                ucs_yaml = demo_yaml
                
                ucs_yaml['name'] = name
                ucs_yaml['location'] = location
                ucs_yaml['owner']['email'] = email
                ucs_yaml['management']['ip-address'] = host_ip
                ucs_yaml['management']['gateway'] = gateway 
                ucs_yaml['management']['netmask-bits']= netmask
                
                ucs_yaml['tags'][1]['tag'] = "Cores - " + str(tags[0])
                ucs_yaml['tags'][2]['tag'] = "memory - " + str(tags[1])
                ucs_yaml['tags'][3]['tag'] = tags[2]
                ucs_yaml['misc'] = ''
                ucs_yaml['tftp'] = ''
                df_peer = rack_info[rack_info['Peer']==peer]
                ucs_yaml["connections"] = []
                
                
                
                for i in range(len(df_peer)):
                    interface = df_peer.iloc[i]["Interface.1"]
                    leaf = df_peer.iloc[i][" Leaf Node# and Intf"]
                    temp_dict={interface:{'eng-blr-aci-spine1':leaf}}
                    ucs_yaml["connections"].insert(i,temp_dict)
                    if i<4:
                        ucs_yaml["cards"][1]["ports"][i]["name"] = interface
                    else:
                        ucs_yaml["cards"][2]["ports"][i-4]["name"] = interface
                
                
              
                yaml_filename = "dev_ENG-BGL-POOL-{}.yaml".format(peer)
                with open(yaml_filename, 'w') as file:
                    yaml.dump(ucs_yaml, file,default_flow_style=False)
                    #print("ucs_yaml",ucs_yaml)
            except:
                print("Router not found")

laas_yaml()
```
