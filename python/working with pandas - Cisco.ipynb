```python
Task

Create a user defined function to find the range of a series

Use the function for aggregation using groupby
```

```python
def cal_range(x):
    
    return x.max() - x.min() 
```

```python
cal_range(cars.Horsepower)
```

```python
titanic.groupby(['Pclass','Sex'])[['Age','Pclass']].agg({'Age':[cal_range],'Pclass':['count']})
```

```python
titanic_groupby = titanic.groupby(['Pclass','Sex'])[['Age','Pclass']].agg({'Age':[cal_range],'Pclass':['count']})
```
